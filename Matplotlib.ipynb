```python
import numpy as np
#from matplotlib import pyplot as plt
x = np.linspace(0,10,10)
y = np.linspace(0,10,10)
plt1, = plt.plot(x,y)
plt2, = plt.plot(x,x**2)
plt3, = plt.plot(x,x**1/2)
#plt.title('Multiple Plots')
#plt.xlabel('x-axis')
#plt.ylabel('y-axis')
(plt.xticks([1,3,5,7,9,11]))
plt.ylim(-10,110)
plt.yticks([-10,10,30,50,70,90,110])
plt.legend([plt1,plt2,plt3],['x','x**2','x**1/2'])
plt.show()
```
